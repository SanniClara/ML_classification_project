```python
training_dataset = df.sample(frac=0.8,random_state=0)
test_dataset = df.drop(training_dataset.index)
```

```python
training_labels = training_dataset.pop('quality')
test_labels = test_dataset.pop('quality')
```

```python
normalized_training_data = normalize(training_dataset)
normalized_test_data = normalize(test_dataset)
```

```python
print(normalized_training_data.shape)
```

```python
# Try adding more Dense layers between these two with activation='relu', differing number of neurons

model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(training_dataset.keys())]),
    layers.Dense(128, activation='relu', input_shape=[len(training_dataset.keys())]),
    layers.Dense(128, activation='relu', input_shape=[len(training_dataset.keys())]),
    layers.Dense(100, activation='relu', input_shape=[len(training_dataset.keys())]),
    layers.Dense(10, activation='softmax', input_shape=[len(training_dataset.keys())]),
    layers.Dense(1)
  ])

model.compile(loss='mse',
              optimizer='adam',
              metrics=['mae', 'mse', 'accuracy'])
```

```python
len(training_dataset.keys())
```

```python
print(model.summary())
```

```python
print(normalized_training_data)
print(normalized_training_data.shape)
```

```python
EPOCHS = 1000 # changed from 1000 to 100

history = model.fit(
  normalized_training_data, training_labels,
  epochs=EPOCHS, validation_split = 0.0005, verbose=0, #changed the spilt from 0.2 to 0.0005
  callbacks=[tfdocs.modeling.EpochDots()])
```
